```python
import numpy as np
import time
import psutil
import matplotlib.pyplot as plt
import tracemalloc
```

## Branch and Bound Approach

```python
def partition_values_from_index(values, start_index, total_value, unassigned_value, test_assignment, test_value, best_assignment, best_err):
    if start_index >= len(values):
        test_err = abs(2 * test_value - total_value)
        if test_err < best_err[0]:
            best_err[0] = test_err
            best_assignment[0] = test_assignment[:]
    else:
        test_err = abs(2 * test_value - total_value)
        if test_err - unassigned_value < best_err[0]:
            unassigned_value -= values[start_index]

            # Try adding values[start_index] to set 1.
            test_assignment[start_index] = True
            partition_values_from_index(values, start_index + 1,
                                        total_value, unassigned_value,
                                        test_assignment, test_value + values[start_index],
                                        best_assignment, best_err)

            # Try adding values[start_index] to set 2.
            test_assignment[start_index] = False
            partition_values_from_index(values, start_index + 1,
                                        total_value, unassigned_value,
                                        test_assignment, test_value,
                                        best_assignment, best_err)
```

```python
def measure_time_and_memory_bnb(values):
    start_time = time.time()
    tracemalloc.start()

    start_index = 0
    total_value = sum(values)
    unassigned_value = total_value
    test_assignment = [False] * len(values)
    test_value = 0
    best_assignment = [None]
    best_err = [float('inf')]

    partition_values_from_index(values, start_index, total_value, unassigned_value, test_assignment, test_value, best_assignment, best_err)
    # print("n = ", len(values))
    # print(best_assignment[0])
    # print(best_err[0])
    # print()

    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    end_time = time.time()
    execution_time = end_time - start_time

    return execution_time, peak
```

## Dynamic Programming Approach

```python
def findPartition(arr, n):
    sum = 0
    i, j = 0, 0

    # calculate sum of all elements
    for i in range(n):
        sum += arr[i]

    if sum % 2 != 0:
        return False

    part = [[True for i in range(n + 1)]
            for j in range(sum // 2 + 1)]

    # initialize top row as true
    for i in range(0, n + 1):
        part[0][i] = True

    # initialize leftmost column,
    # except part[0][0], as 0
    for i in range(1, sum // 2 + 1):
        part[i][0] = False

    # fill the partition table in
    # bottom up manner
    for i in range(1, sum // 2 + 1):

        for j in range(1, n + 1):
            part[i][j] = part[i][j - 1]

            if i >= arr[j - 1]:
                part[i][j] = (part[i][j] or
                                part[i - arr[j - 1]][j - 1])

    return part[sum // 2][n]
```

```python
def measure_time_and_memory_dp(values):
    start_time = time.time()
    tracemalloc.start()


    n = len(values)
    findPartition(values, n)

    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    end_time = time.time()
    execution_time = end_time - start_time

    return execution_time, peak
```

```python
sizes = [20, 40, 80]
times_bnb = []
mem_usages_bnb = []
times_dp = []
mem_usages_dp = []
values_arr = []

for size in sizes:
    values = np.random.randint(0, 100, size)
    values_arr.append(values)
    execution_time, memory_usage = measure_time_and_memory_bnb(values)
    times_bnb.append(execution_time)
    mem_usages_bnb.append(memory_usage)
    execution_time, memory_usage = measure_time_and_memory_dp(values)
    times_dp.append(execution_time)
    mem_usages_dp.append(memory_usage)
```

```python
# Save values in values_arr in .txt file
with open('values.txt', 'w') as f:
    for item in values_arr:
        f.write("%s\n" % item)
```
